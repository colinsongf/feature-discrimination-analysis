```python
home_dir = '/Users/ianballard/Dropbox/fd/melodic/'
if not os.path.exists(home_dir):
    os.mkdir(home_dir)

subj_file = '/Users/ianballard/Dropbox/fd/scripts/subjects_ser.txt'
subjects = list(np.loadtxt(subj_file,'string'))

runs = ['1','2','3']
exp = ['sim','ser']

for sub in subjects:
    sub_dir = home_dir + sub + '/'
    if not os.path.exists(sub_dir):
        os.mkdir(sub_dir)
    
    for e in exp:
        exp_dir = sub_dir + e + '/'
        if not os.path.exists(exp_dir):
            os.mkdir(exp_dir)   
            
        for run in runs:
            run_dir = exp_dir + 'run_' + run + '/'
            if not os.path.exists(run_dir):
                os.mkdir(run_dir) 
```

```python
in_tuples = []
for sub in subjects:
    for e in exp:
        for run in runs:
            in_tuples.append((sub,e,run))
pool = multiprocessing.Pool(processes = 12)
pool.map(copy_melodic,in_tuples)
```
